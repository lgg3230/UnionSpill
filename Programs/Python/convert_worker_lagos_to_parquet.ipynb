### Convert worker lagos sample to parquet:

```python
import pandas as pd, duckdb
import numpy as np
import pyarrow as pa
import pyarrow.parquet as pq
import os
import polars as pl
```

```python
df = pd.read_stata("/kellogg/proj/lgg3230/UnionSpill/Data/RAIS_aux/worker_estab_lagos.dta", convert_categoricals=False)
```

```python
table = pa.Table.from_pandas(df)
pq.write_table(table, "/kellogg/proj/lgg3230/UnionSpill/Data/RAIS_aux/worker_estab_lagos.parquet")
```
